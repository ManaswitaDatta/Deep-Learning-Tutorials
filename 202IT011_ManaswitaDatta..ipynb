To define the functions for performing the linear regression on the dataset that is imported, a few functions need to be defined first. 

### Gradient Partial Derivatives Definition
To run gradient descent on the error function, the gradient need to be computed beforehand. The gradient will act like a compass and always point downhill. To compute it, the error function is differentiate. Since the function is defined by two
parameters i.e. m and c, two partial derivative will be computed, one for each, but both in the same function definition.

```python
# Performing Gradient Descent 
def grad_descent(L, epochs):
    m = 0
    c = 0
    n = float(len(X)) # Number of elements in X
    D_m = [0]*epochs
    D_c = [0]*epochs
    # Performing Gradient Descent 
    for i in range(epochs): 
        Y_pred = m*X + c  # The current predicted value of Y
        D_m[i] = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
        D_c[i] = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
        m = m - L * D_m[i]  # Update m
        c = c - L * D_c[i]  # Update c

    print (m, c)
    return m, c, D_m, D_c
```

### Error Function Plotting

The purpose of this function is to plot the slope (m) and intercept (c) error calculated from the previous gradient descent function and see how that behave over the course of a certain number of iterations 

```python
def plot_error(error):
    
    plt.scatter(np.linspace(0,len(error),len(error)), error, alpha=0.5)
    plt.xlabel('epochs')
    plt.grid()
    plt.show()
```

### Calculation Regression over 500 Iterations with 0.01 learning rate

```python
L = 0.01  # The learning Rate
epochs = 500  # The number of iterations to perform gradient descent
m, c, m_error, c_error = grad_descent(L, epochs)  
Y_pred = m*X + c
plt.scatter(X, Y)
plt.ylabel('Min temp')
plt.xlabel('Max temp')
plt.title('Gradient descent\n L = 0.01, epochs = 500')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') 
plt.show()

plt.title('m Error\n L = 0.01, epochs = 500')
plt.ylabel('m error')
plot_error(m_error)
plt.title('c Error\n L = 0.01, epochs = 500')
plt.ylabel('c error')
plot_error(c_error)
```

The value of m and c is nan, nan respectively. To solve that, the learning rate should be decreased. The conclusion is that gradient descent is sensitive to learning rate. The value of m_error and c_error is 0, 0 respectively as expected.

### Calculation Regression over 500 Iterations with 0.001 learning rate

```python
L = 0.001  # The learning Rate
epochs = 500  # The number of iterations to perform gradient descent
m, c, m_error, c_error = grad_descent(L, epochs)  
Y_pred = m*X + c
plt.scatter(X, Y)
plt.ylabel('Min temp')
plt.xlabel('Max temp')
plt.title('Gradient descent\n L = 0.001, epochs = 500')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') 
plt.show()

plt.title('m Error\n L = 0.001, epochs = 500')
plt.ylabel('m error')
plot_error(m_error)
plt.title('c Error\n L = 0.001, epochs = 500')
plt.ylabel('c error')
plot_error(c_error)
```

Still the value of m and c is 0.724, 0.125 respectively. The m error and c error converged around 10 - 20 iterations.

### Calculation Regression over 500 Iterations with 0.0001 learning rate

```python
L = 0.0001  # The learning Rate
epochs = 500  # The number of iterations to perform gradient descent
m, c, m_error, c_error = grad_descent(L, epochs)  
Y_pred = m*X + c
plt.scatter(X, Y)
plt.ylabel('Min temp')
plt.xlabel('Max temp')
plt.title('Gradient descent\n L = 0.0001, epochs = 500')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') 
plt.show()

plt.title('m Error\n L = 0.0001, epochs = 500')
plt.ylabel('m error')
plot_error(m_error)
plt.title('c Error\n L = 0.0001, epochs = 500')
plt.ylabel('c error')
plot_error(c_error)
```

The value of m and c is 0.727, 0.034 respectively. 
From the plots above it can be seen that the value for m error and c error converged somewhere close to 20-30 iterations. 

### Calculation Regression over 500 Iterations with 0.00001 learning rate

```python
L = 0.00001  # The learning Rate
epochs = 500  # The number of iterations to perform gradient descent
m, c, m_error, c_error = grad_descent(L, epochs)  
Y_pred = m*X + c
plt.scatter(X, Y)
plt.ylabel('Min temp')
plt.xlabel('Max temp')
plt.title('Gradient descent\n L = 0.00001, epochs = 500')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') 
plt.show()

plt.title('m Error\n alpha = 0.00001, epochs = 500')
plt.ylabel('m error')
plot_error(m_error)
plt.title('c Error\n alpha = 0.00001, epochs = 500')
plt.ylabel('c error')
plot_error(c_error)
```

We can see that the m_error and c_error has coverged in 220 - 230 iterations. These plots give us the confidence to say that our line of best fit would have the final equation of y = 0.724x + 0.125 after performing the gradient descent algorithm.

The effect of change in learning rate from 0.01 to 0.00001 is seen using the gradient descent algorithm.
